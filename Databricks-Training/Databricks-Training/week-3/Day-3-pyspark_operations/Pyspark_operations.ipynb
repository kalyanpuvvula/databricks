## select operations

```python

df.select("emp_id","emp_name","department").show()
df.filter(col("department")=="IT").select("emp_id","emp_name","salary","department").show()
df.select("emp_id","emp_name","salary","department").limit(5).show() 
df.filter(col("city")=="Hyderabad").show()
df.drop("joining_date").show()
```

##Alias 

```python
df.select(
    col("joining_date").alias("obj"),
    col("emp_name").alias("employee_name"),
    col("designation").alias("job_role"),
    col("salary").alias("montly_salary")
).display()
```

## Filter/Where

```python
df.filter(col("salary")>70000).limit(5).show()
df.filter(col("city")=="Hyderabad").show()
df.filter(col("age")>25).limit(5).show()
df.filter(col("emp_name").startswith("s")).show()
df.filter(col("emp_name").endswith("a")).show()
df.filter(col("designation").contains("Engineer")).show()
df.filter(col("joining_date") > "2022-01-01").show()
```

## withColumnRenamed

```python
df = df.withColumnRenamed("emp_name", "employee_name") \
       .withColumnRenamed("salary", "monthly_salary") \
       .withColumnRenamed("city", "worklocation") \
       .withColumnRenamed("department", "dept").show()

```

##columnwith

```python
df = spark.createDataFrame(data, columns)

df.show()
```

```python
# Bonus column
df.withColumn("bonus", col("salary") * 0.10).show()

# Annual salary
df.withColumn("annual_salary", col("salary") * 12).show()

# Tax column
df.withColumn("tax", col("salary") * 0.05).show()

# Updated salary
df.withColumn("updated_salary", col("salary") + 5000).show()

# Salary category
df.withColumn(
    "salary_category",
    when(col("salary") >= 80000, "High")
    .when(col("salary") >= 50000, "Medium")
    .otherwise("Low")
).show()

# Age group
df.withColumn(
    "age_group",
    when(col("age") < 25, "Young")
    .otherwise("Adult")
).show()

# Location column
df.withColumn(
    "location",
    concat(col("city"), lit("-"), col("department"))
).show()

# Name length
df.withColumn(
    "name_length",
    length(col("emp_name"))
).show()

# Joining year
df.withColumn(
    "joining_year",
    year(col("joining_date"))
).show()
```
