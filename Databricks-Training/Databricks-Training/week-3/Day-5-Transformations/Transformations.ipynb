##26. isin() Questions

```python
emp_df.filter(col("city").isin("Hyderabad","Bangalore")).show()

emp_df.filter(col("department").isin("IT","QA")).show()

emp_df.filter(col("age").isin(28,30,35)).show()

emp_df.filter(
    (col("gender")=="Female") &
    (col("city").isin("Pune","Chennai"))
).show()

emp_df.filter(col("emp_id").isin(101,104,108)).show()
```

##27. between() Questions

```python
emp_df.filter(col("salary").between(50000,80000)).show()

emp_df.filter(col("age").between(25,30)).show()

emp_df.filter(col("salary").between(60000,90000)).show()

emp_df.filter(
    year(to_date("joining_date"))
    .between(2020,2022)
).show()

emp_df.filter(col("emp_id").between(102,108)).show()
```

##28. pivot()

```python
emp_df.groupBy().pivot("department").avg("salary").show()

emp_df.groupBy().pivot("city").count().show()

emp_df.groupBy().pivot("gender").sum("salary").show()

emp_df.groupBy().pivot("department").max("age").show()

emp_df.groupBy("department").agg(
    count("*").alias("employee_count"),
    avg("salary").alias("avg_salary")
).show()
```

##29. unpivot / stack()

##30. Window Functions

```python
from pyspark.sql.window import Window

w = Window.orderBy(col("salary").desc())

emp_df.withColumn(
    "rank",
    rank().over(w)
).show()

w2 = Window.partitionBy("department")\
           .orderBy(col("salary").desc())

emp_df.withColumn(
    "dense_rank",
    dense_rank().over(w2)
).show()

emp_df.withColumn(
    "row_number",
    row_number().over(w)
).show()

emp_df.withColumn(
    "lead_salary",
    lead("salary").over(w)
).show()

emp_df.withColumn(
    "lag_salary",
    lag("salary").over(w)
).show()
```

##42. zipWithIndex()

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number, monotonically_increasing_id

w = Window.orderBy("emp_id")

# 1. Add index
emp_df.withColumn(
    "index",
    row_number().over(w)
).show()

# 2. Generate sequence numbers
emp_df.withColumn(
    "sequence_no",
    row_number().over(w)
).show()

# 3. Create row IDs
emp_df.withColumn(
    "row_id",
    row_number().over(w)
).show()

# 4. Compare with monotonically_increasing_id
emp_df.withColumn(
    "mono_id",
    monotonically_increasing_id()
).show()

# 5. Indexed employee list
emp_df.withColumn(
    "employee_index",
    row_number().over(w)
).select(
    "employee_index",
    "name"
).show()
```
